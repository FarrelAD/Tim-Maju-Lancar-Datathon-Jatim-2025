# Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from pathlib import Path
```

# SLB Students Year by Year

## Jawa Timur

```python
jatim_csv_files = glob.glob('../data/dapo-kemdikdasmen/*/Prov. Jawa Timur/ALL/jatim-csv.csv')

df_jatim_list = []

for file in jatim_csv_files:
    df = pd.read_csv(file)
    
    df['semester_id'] = Path(file).parent.parent.parent.stem
    
    df_jatim_list.append(df)

df_slb_jatim_yby = pd.concat(df_jatim_list, ignore_index=True)
df_slb_jatim_yby
```

```python
summed_by_semester_id = df_slb_jatim_yby.groupby('semester_id').sum(numeric_only=True)
summed_by_semester_id.to_csv('../outputs/data-jatim-group-by-semester-id.csv')
summed_by_semester_id
```

### Total Students Year by Year

```python
summed_by_semester_id['pd'].plot(kind='line', marker='o', legend=False)
plt.title("Total Siswa SLB dari tahun ke tahun")
plt.xlabel("Semester")
plt.ylabel("Total Siswa SLB")
plt.grid(True)
plt.show()
```

### Total Teachers Year by Year

```python
summed_by_semester_id['ptk'].plot(kind='line', marker='o', legend=False)
plt.title("Total Guru SLB dari tahun ke tahun")
plt.xlabel("Semester")
plt.ylabel("Total Guru SLB")
plt.grid(True)
plt.show()
```

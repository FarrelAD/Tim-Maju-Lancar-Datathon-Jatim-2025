# Preparation

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import glob
from pathlib import Path
```

# SLB Students Year by Year

## Jawa Timur

```python
jatim_csv_files = glob.glob('../data/dapo-kemdikdasmen/*/Prov. Jawa Timur/ALL/jatim-csv.csv')

df_jatim_list = []

for file in jatim_csv_files:
    df = pd.read_csv(file)
    
    df['semester_id'] = Path(file).parent.parent.parent.stem
    
    df_jatim_list.append(df)

df_slb_jatim_yby = pd.concat(df_jatim_list, ignore_index=True)
df_slb_jatim_yby
```

```python
summed_by_semester_id = df_slb_jatim_yby.groupby('semester_id').sum(numeric_only=True)
summed_by_semester_id.to_csv('../outputs/data-jatim-group-by-semester-id.csv')
summed_by_semester_id
```

### Total Students Year by Year

```python
# calculate year-over-year percentage change `pd` and `ptk``
summed_by_semester_id['persentase_kenaikan_pd'] = summed_by_semester_id['pd'].pct_change() * 100
summed_by_semester_id['persentase_kenaikan_ptk'] = summed_by_semester_id['ptk'].pct_change() * 100
summed_by_semester_id
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Left y-axis → total students
ax1.plot(
    summed_by_semester_id.index,
    summed_by_semester_id['pd'],
    marker='o',
    color='blue',
    label='Total Siswa SLB'
)
ax1.set_xlabel("Semester")
ax1.set_ylabel("Total Siswa SLB", color='blue')
ax1.tick_params(axis='y', labelcolor='blue')
ax1.grid(True)

# Right y-axis → percentage increase
ax2 = ax1.twinx()
ax2.plot(
    summed_by_semester_id.index,
    summed_by_semester_id['persentase_kenaikan_pd'],
    marker='s',
    linestyle='--',
    color='red',
    label='Persentase Kenaikan'
)
ax2.set_ylabel("Persentase Kenaikan (%)", color='red')
ax2.tick_params(axis='y', labelcolor='red')

# Title
plt.title("Total Siswa SLB dan Persentase Kenaikan dari Tahun ke Tahun")

# Legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

plt.show()
```

### Total Teachers Year by Year

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

# Left y-axis → total teachers
ax1.plot(
    summed_by_semester_id.index,
    summed_by_semester_id['ptk'],
    marker='o',
    color='green',
    label='Total Guru SLB'
)
ax1.set_xlabel("Semester")
ax1.set_ylabel("Total Guru SLB", color='green')
ax1.tick_params(axis='y', labelcolor='green')
ax1.grid(True)

# Right y-axis → percentage increase
ax2 = ax1.twinx()
ax2.plot(
    summed_by_semester_id.index,
    summed_by_semester_id['persentase_kenaikan_ptk'],
    marker='s',
    linestyle='--',
    color='orange',
    label='Persentase Kenaikan Guru'
)
ax2.set_ylabel("Persentase Kenaikan (%)", color='orange')
ax2.tick_params(axis='y', labelcolor='orange')

# Title
plt.title("Total Guru SLB dan Persentase Kenaikan dari Tahun ke Tahun")

# Legends
lines, labels = ax1.get_legend_handles_labels()
lines2, labels2 = ax2.get_legend_handles_labels()
ax1.legend(lines + lines2, labels + labels2, loc="upper left")

plt.show()
```
